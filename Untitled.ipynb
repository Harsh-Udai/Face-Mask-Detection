```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data = cv2.CascadeClassifier('haarcascade.xml')
```

```python
# Code for without mask data capturing
capture = cv2.VideoCapture(0)
face_data=[]
while True:
    flag, img = capture.read()
    if flag:
        faces = data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            
            print(len(face_data))
            if len(face_data)<400:
                face_data.append(face)
            
        cv2.imshow('result',img)
        
        if cv2.waitKey(2)==27 or len(face_data) >=200:
            break
            
capture.release()
cv2.destroyAllWindows()
```

```python
np.save("without_mask.npy",face_data)
```

```python
# Code for with mask data capturing
capture = cv2.VideoCapture(0)
face_data=[]
while True:
    flag, img = capture.read()
    if flag:
        faces = data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            
            print(len(face_data))
            if len(face_data)<400:
                face_data.append(face)
            
        cv2.imshow('result',img)
        
        if cv2.waitKey(2)==27 or len(face_data) >=200:
            break
            
capture.release()
cv2.destroyAllWindows()
```

```python
np.save("with_mask.npy",face_data)
```

```python
# After Work
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
# Converting 4 dimension to 2 dimension
with_mask = with_mask.reshape(200,50*50*3)
without_mask = without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
# Concatenation of both the data

X = np.r_[with_mask, without_mask]
```

```python
X.shape
```

```python
# Creating labels first 200 are with mask are assigned as 0 and remaining as 1.
labels = np.zeros(X.shape[0])
```

```python
labels[200:] = 1.0
```

```python
# Now Machine Learning Part
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,labels, test_size=0.20)
```

```python
x_train.shape
```

```python
# Now doing dimensionality reduction
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
x_train = pca.fit_transform(x_train)
```

```python
x_train.shape
```

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
names = {0:'Mask', 1:'No Mask'}
```

```python
# Now doing some real time stuff
capture = cv2.VideoCapture(0)

font = cv2.FONT_HERSHEY_COMPLEX

while True:
    flag, img = capture.read()
    if flag:
        faces = data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            face = face.reshape(1,-1)
            face = pca.transform(face)
            pred = svm.predict(face)
            n = names[int(pred)]
            print(n)
            cv2.putText(img, n, (x,y), font, 1, (244,250,250), 2)
    
        cv2.imshow('result',img)
        
        if cv2.waitKey(2)==27:
            break
            
capture.release()
cv2.destroyAllWindows()
```
